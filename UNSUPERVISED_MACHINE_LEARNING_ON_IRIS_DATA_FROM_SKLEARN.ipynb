```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.preprocessing import StandardScaler

# Standardize features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df)

# Convert the scaled data back to a DataFrame
df_scaled = pd.DataFrame(scaled_features, columns=df.columns)
```

```python
# Determine the optimal number of clusters using the Elbow Method
inertia = []
k_range = range(1, 11)  # Testing cluster numbers from 1 to 10

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(df_scaled)
    inertia.append(kmeans.inertia_)

# Plot the Elbow Method result
plt.figure(figsize=(8, 6))
plt.plot(k_range, inertia, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.xticks(k_range)
plt.grid(True)
plt.show()
```

KMEANS CLUSTERING

K-Means clustering is a popular unsupervised learning algorithm used to partition data into distinct clusters based on feature similarity. Applying K-Means to the Iris dataset groups flowers into clusters based on sepal and petal measurements. The number of clusters is determined using the Elbow method, which suggests k=3 as the optimal number. This division ideally corresponds to the three species of iris. Standardizing the features and applying K-Means reveal patterns in the data, which can be visualized using scatter plots

```python
from sklearn.cluster import KMeans

# Assuming optimal number of clusters is found to be `n_clusters`
n_clusters = 3  # Replace with your optimal number of clusters

# Apply K-Means clustering
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
df['cluster'] = kmeans.fit_predict(df_scaled)
df
```

```python
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='cluster', palette='summer', data=df)
plt.title('K-Means Clustering')
plt.show()
```

**Hierarchical clustering**

Hierarchical clustering on the Iris dataset, specifically using agglomerative hierarchical clustering, organizes flowers into a hierarchy of clusters based on sepal and petal measurements. The algorithm starts with each flower as an individual cluster and iteratively merges the closest clusters. The resulting dendrogram shows how clusters combine at various levels of similarity, helping to identify the optimal number of clusters by analyzing the tree structure.

```python
from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering(n_clusters = 3, affinity='euclidean', linkage='ward')
# Assuming optimal number of clusters is found to be `n_clusters`
n_clusters = 3  # Replace with your optimal number of clusters

df['cluster'] = model.fit_predict(df_scaled)
df
```

```python
feature_pairs = [
    ('sepal length (cm)', 'sepal width (cm)'),
    ('petal length (cm)', 'petal width (cm)'),
    ('sepal length (cm)', 'petal length (cm)'),
    ('sepal width (cm)', 'petal width (cm)')
]

# Create scatter plots for each feature pair
for x_feature, y_feature in feature_pairs:
    plt.figure(figsize=(10, 8))
    sns.scatterplot(x=x_feature, y=y_feature, hue='cluster', palette='summer', data=df)
    plt.title(f'A Clustering: {x_feature} vs {y_feature}')
    plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
sc = linkage(df, method = 'ward')
dendrogram(sc, leaf_rotation=90)
import matplotlib.pyplot as plt
plt.xlabel('Data Points')
plt.ylabel('Euclidean Distance')
plt.tight_layout()
plt.title('Dendrogram')
plt.show()
```
